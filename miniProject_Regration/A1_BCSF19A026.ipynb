```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "drive/MyDrive/assignment1/dataset/"
train_df = pd.read_csv(path+"train.csv")
test_df = pd.read_csv(path+"test.csv")
print(test_df.head(10))
print(train_df.head(10))
```

```python
train_arr = train_df.to_numpy()
test_arr = test_df.to_numpy()
print("Shape of arr1:", train_arr.shape)
print("Shape of arr2:", test_arr.shape)
```

```python
cost_min = float('inf')
opt_theta0 = None
opt_theta1= None
m=len(train_arr)
theta0=-0.5
theta1=-0.5
#finding hypotheses func and cost
def find_cost(x, y, theta0, theta1):
   #print(x,y,theta0, theta1)
   hypo_theses = theta0 + theta1 * x  
   #print(hypo_theses) 
   cost = np.sum((hypo_theses - y) ** 2) / (2 * m)
   return cost

for theta0 in np.arange(-0.5,1.5,0.2):
    for theta1 in np.arange(-0.5,1.5,0.2):
        cost = (find_cost(train_arr[:,0],train_arr[:,1],theta0,theta1))
        #print(cost)
        if cost < cost_min:
          cost_min = cost
          opt_theta0 = theta0
          opt_theta1 = theta1                                                                                                                                     # Min Cost


print("Minimum cost ", cost_min)
print("Best theta0 is ", opt_theta0)
print("Best theta1 is ", opt_theta1)
th0 = opt_theta0
th1 = opt_theta1



```

```python
test_preout = []
for i in range(len(test_arr)):
    hypo_theses = th0 + th1 * test_arr[i][0]
    test_preout.append(hypo_theses)
for i in range(len(test_arr)):
  print("Predicted output",test_preout[i])
  print("Actual output" ,test_arr[i][1])
```

```python
m = len(test_preout)
error = 0
outputs=test_arr[:,1]
for i in range(len(test_preout)):
    error += (test_preout[i] - outputs[i]) ** 2
error=error / (2 * m)
print(error)
```

```python
test=test_arr[:,0]
plt.plot(test, outputs, color='Red')
plt.plot(test, test_preout, color='Blue')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
